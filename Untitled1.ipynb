```python
# ==========================================
# BOOK REVIEW SCRAPER
# Website: books.toscrape.com
# ==========================================

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

BASE_URL = "https://books.toscrape.com/"

# ------------------------------------------
# Convert rating text to number
# ------------------------------------------
def get_rating(rating_class):

    ratings = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }

    return ratings.get(rating_class, 0)


# ------------------------------------------
# Scrape books from one page
# ------------------------------------------
def scrape_page(url):

    response = requests.get(url)

    if response.status_code != 200:
        print("Failed to fetch:", url)
        return []

    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")

    data = []

    for book in books:

        # Book Name
        title = book.h3.a["title"]

        # Rating
        rating_tag = book.find("p", class_="star-rating")

        rating_class = rating_tag["class"][1]

        rating = get_rating(rating_class)

        # Price
        price = book.find("p", class_="price_color").text.strip()

        # Availability
        availability = book.find(
            "p",
            class_="instock availability"
        ).text.strip()

        # Fake placeholders because this website
        # does NOT contain real comments/reviews/users
        comment = "No real comments available"
        customer_name = "Anonymous"
        comment_tags = ["book", "rating"]

        data.append({
            "Title": title,
            "Rating": rating,
            "Price": price,
            "Availability": availability,
            "Customer": customer_name,
            "Comment": comment,
            "Tags": ", ".join(comment_tags)
        })

    return data


# ------------------------------------------
# Scrape multiple pages
# ------------------------------------------
all_books = []

for page in range(1, 6):

    if page == 1:
        url = BASE_URL
    else:
        url = BASE_URL + f"catalogue/page-{page}.html"

    print(f"Scraping Page {page}...")

    books = scrape_page(url)

    all_books.extend(books)

    time.sleep(1)


# ------------------------------------------
# Save data into CSV
# ------------------------------------------
df = pd.DataFrame(all_books)

df.to_csv("books_reviews.csv", index=False)

print("\nScraping Completed")
print(df.head())
```
