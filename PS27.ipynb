```python
Create a review scrapper for https://store.steampowered.com/appreviews
/730?json=1 website to fetch real time comments, reviews, ratings, comment
tags, customer name using Python.
```

```python
import requests
import pandas as pd
```

```python
url = "https://store.steampowered.com/appreviews/730?json=1"
```

```python
#Fetch Data from Website
response = requests.get(url)
data = response.json()
```

```python
#Check API Data
print(data.keys())
```

```python
#Extract Reviews
reviews = data['reviews']
print("Total reviews fetched:", len(reviews))
```

```python
# Extract Required Information

# Now we will extract:

# Customer Name
# Review Comment
# Rating
# Tags
# Playtime

review_data = []

for review in reviews:

    customer_name = review['author']['steamid']
    comment = review['review']
    rating = "Positive" if review['voted_up'] else "Negative"
    playtime = review['author']['playtime_forever']
    tags = review.get('weighted_vote_score', 0)

    review_data.append({
    "Customer Name": customer_name,
    "Comment": comment,
    "Rating": rating,
    "Playtime": playtime,
    "Tag Score": tags
    })
```

```python
#Convert Data into Table
df = pd.DataFrame(review_data)
```

```python
df.to_csv("data.csv", index= False)
print("generated csv")
```
