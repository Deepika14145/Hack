```python
Perform the following operations using Python on ForestFires Dataset.
a. Create data subsets by making classesfor amount of region affected.(e.g.
NotAffected, Partially affected, Mostlyaffected).
b. Merge two subsets
c. Sort Data using Temperature, wind and area.
d. Transposing Data
e. Melting Data to long format
f. Casting data to wide format
```

```python
# =========================================================
# FOREST FIRES DATASET OPERATIONS
# =========================================================

# =========================================================
# STEP 1 : IMPORT LIBRARIES
# =========================================================

import pandas as pd


# =========================================================
# STEP 2 : LOAD DATASET
# =========================================================

data = pd.read_csv("forestfires.csv")

# Display first 5 rows

print("FIRST 5 ROWS")
print(data.head())


# =========================================================
# a. CREATE DATA SUBSETS
# =========================================================
# Create classes based on area affected
# =========================================================

# NotAffected      -> area = 0
# PartiallyAffected -> area between 0 and 50
# MostlyAffected   -> area > 50

not_affected = data[data['area'] == 0]

partially_affected = data[
    (data['area'] > 0) &
    (data['area'] <= 50)
]

mostly_affected = data[data['area'] > 50]

print("\nNOT AFFECTED")
print(not_affected.shape)

print("\nPARTIALLY AFFECTED")
print(partially_affected.shape)

print("\nMOSTLY AFFECTED")
print(mostly_affected.shape)


# =========================================================
# b. MERGE TWO SUBSETS
# =========================================================
# Merge NotAffected and PartiallyAffected
# =========================================================

merged_data = pd.concat(
    [not_affected, partially_affected]
)

print("\nMERGED DATA")
print(merged_data.shape)


# =========================================================
# c. SORT DATA
# =========================================================
# Sort using temperature, wind and area
# =========================================================

sorted_data = data.sort_values(
    by=['temp', 'wind', 'area']
)

print("\nSORTED DATA")
print(sorted_data.head())


# =========================================================
# d. TRANSPOSE DATA
# =========================================================
# Convert rows into columns
# =========================================================

transposed_data = data.transpose()

print("\nTRANSPOSED DATA")
print(transposed_data.head())


# =========================================================
# e. MELT DATA TO LONG FORMAT
# =========================================================
# Convert wide format into long format
# =========================================================

melted_data = pd.melt(

    data,

    id_vars=['month', 'day'],

    value_vars=['temp', 'wind', 'RH'],

    var_name='Weather_Type',

    value_name='Value'
)

print("\nMELTED DATA")
print(melted_data.head())


# =========================================================
# f. CAST DATA TO WIDE FORMAT
# =========================================================
# Convert long format back to wide format
# =========================================================

wide_data = melted_data.pivot_table(

    index=['month', 'day'],

    columns='Weather_Type',

    values='Value',

    aggfunc='mean'
)

print("\nWIDE FORMAT DATA")
print(wide_data.head())


# =========================================================
# END
# =========================================================
```
