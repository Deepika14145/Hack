```python
#Create a review scrapper for https://webscraper.io/test-sites/e-commerce
/static website to fetch realtime comments, reviews, ratings, comment tags,
customer name using Python.
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/static"
```

```python
response = requests.get(url)
html_content = response.text
```

```python
#Parse HTML
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
#Find All Products
products = soup.find_all("div", class_="thumbnail")
print("Total products: ", len(products))
```

```python

#Convert HTML Data into JSON Format all_products = []
product_data = []

for product in products:

    # Product Name
    name = product.find('a', class_='title').text.strip()

    # Price
    price = product.find('h4', class_='price').text.strip()

    # Description
    description = product.find('p', class_='description').text.strip()

# Reviews Count
review_tag = product.find('p', class_='pull-right')

if review_tag:
    reviews = review_tag.text.strip()
else:
    reviews = "No Reviews"
    # Ratings
    rating = len(product.find_all('span', class_='glyphicon-star'))

    # Tags
    tags = "Electronics Product"

    product_data.append({
        "Customer/Product Name": name,
        "Price": price,
        "Comment/Description": description,
        "Reviews": reviews,
        "Rating": rating,
        "Tags": tags
    })
```

```python
df = pd.DataFrame(product_data)
```

```python
df.to_csv("ecommerce_reviews.csv", index=False)

print("CSV File Saved Successfully")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL
url = "https://webscraper.io/test-sites/e-commerce/static"

# Fetch Website Data
response = requests.get(url)

# Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all product cards
products = soup.find_all('div', class_='thumbnail')

# Store data
product_data = []

for product in products:

    # Product Name
    name = product.find('a', class_='title').text.strip()

    # Price
    price = product.find('h4', class_='price').text.strip()

    # Description
    description = product.find('p', class_='description').text.strip()

    # Reviews
# Reviews Count
review_tag = product.find('p', class_='pull-right')

if review_tag:
    reviews = review_tag.text.strip()
else:
    reviews = "No Reviews"
    
    # Ratings
    rating = len(product.find_all('span', class_='glyphicon-star'))

    # Tags
    tags = "Electronics Product"

    # Append data
    product_data.append({
        "Customer/Product Name": name,
        "Price": price,
        "Comment/Description": description,
        "Reviews": reviews,
        "Rating": rating,
        "Tags": tags
    })

# Convert into DataFrame
df = pd.DataFrame(product_data)

# Display data
print(df.head())

# Save into CSV
df.to_csv("ecommerce_reviews.csv", index=False)

print("Scraping Completed Successfully")
```
