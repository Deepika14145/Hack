```python
Perform the following operations using Python on Hepatitis Dataset.
a. Create data subsetsfor differentsex.
b. Merge two subsets
c. Sort Data using age, SGOT, PROTIME.
d. Transposing Data
e. Melting Data to long format
f. Casting data to wide format
```

```python
# =========================================================
# HEPATITIS DATASET OPERATIONS
# =========================================================

import pandas as pd
import numpy as np

# =========================================================
# COLUMN NAMES
# =========================================================

columns = [

    'CLASS',
    'AGE',
    'SEX',
    'STEROID',
    'ANTIVIRALS',
    'FATIGUE',
    'MALAISE',
    'ANOREXIA',
    'LIVER_BIG',
    'LIVER_FIRM',
    'SPLEEN_PALPABLE',
    'SPIDERS',
    'ASCITES',
    'VARICES',
    'BILIRUBIN',
    'ALK_PHOSPHATE',
    'SGOT',
    'ALBUMIN',
    'PROTIME',
    'HISTOLOGY'

]

# =========================================================
# STEP 1 : LOAD DATASET
# =========================================================

data = pd.read_csv(

    "hepatitis.csv",

    header=None,

    names=columns

)

# =========================================================
# STEP 2 : DISPLAY DATA
# =========================================================

print("FIRST 5 ROWS")
print(data.head())

print("\nCOLUMN NAMES")
print(data.columns)

print("\nDATASET SHAPE")
print(data.shape)

# =========================================================
# STEP 3 : DATA CLEANING
# =========================================================

# Replace ? with NaN

data.replace('?', np.nan, inplace=True)

# Convert all columns to numeric

for column in columns:

    data[column] = pd.to_numeric(

        data[column],

        errors='coerce'

    )

# Remove missing values

data.dropna(inplace=True)

# Remove duplicate rows

data.drop_duplicates(inplace=True)

print("\nDATA AFTER CLEANING")
print(data.shape)

# =========================================================
# a. CREATE DATA SUBSETS FOR DIFFERENT SEX
# =========================================================

print("\nSEX VALUES")
print(data['SEX'].unique())

# 1 = Male

male_data = data[data['SEX'] == 1]

# 2 = Female

female_data = data[data['SEX'] == 2]

print("\nMALE DATA")
print(male_data.head())

print("\nFEMALE DATA")
print(female_data.head())

# =========================================================
# b. MERGE TWO SUBSETS
# =========================================================

merged_data = pd.concat(

    [male_data, female_data],

    ignore_index=True

)

print("\nMERGED DATA")
print(merged_data.head())

print("\nMERGED SHAPE")
print(merged_data.shape)

# =========================================================
# c. SORT DATA USING AGE, SGOT, PROTIME
# =========================================================

sorted_data = data.sort_values(

    by=['AGE', 'SGOT', 'PROTIME']

)

print("\nSORTED DATA")
print(sorted_data.head())

# =========================================================
# d. TRANSPOSE DATA
# =========================================================

transposed_data = data.transpose()

print("\nTRANSPOSED DATA")
print(transposed_data.head())

# =========================================================
# e. MELTING DATA TO LONG FORMAT
# =========================================================

melted_data = pd.melt(

    data,

    id_vars=['CLASS', 'SEX'],

    value_vars=['AGE', 'SGOT', 'PROTIME'],

    var_name='Parameter',

    value_name='Value'

)

print("\nMELTED DATA")
print(melted_data.head())

# =========================================================
# f. CASTING DATA TO WIDE FORMAT
# =========================================================

wide_data = melted_data.pivot_table(

    index=['CLASS', 'SEX'],

    columns='Parameter',

    values='Value',

    aggfunc='mean'

)

print("\nWIDE FORMAT DATA")
print(wide_data.head())

# =========================================================
# END OF PROGRAM
# =========================================================
```

```python
# =========================================================
# HEPATITIS DATASET MACHINE LEARNING PROJECT
# =========================================================
# Operations:
# q. Data Cleaning
# r. Outlier Detection
# s. Data Transformation
# t. Logistic Regression and Naive Bayes
# =========================================================

import pandas as pd
import numpy as np

# Machine Learning Libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Models

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

# Accuracy

from sklearn.metrics import accuracy_score

# =========================================================
# COLUMN NAMES
# =========================================================

columns = [

    'CLASS',
    'AGE',
    'SEX',
    'STEROID',
    'ANTIVIRALS',
    'FATIGUE',
    'MALAISE',
    'ANOREXIA',
    'LIVER_BIG',
    'LIVER_FIRM',
    'SPLEEN_PALPABLE',
    'SPIDERS',
    'ASCITES',
    'VARICES',
    'BILIRUBIN',
    'ALK_PHOSPHATE',
    'SGOT',
    'ALBUMIN',
    'PROTIME',
    'HISTOLOGY'

]

# =========================================================
# STEP 1 : LOAD DATASET
# =========================================================

data = pd.read_csv(

    "hepatitis.csv",

    header=None,

    names=columns

)

print("FIRST 5 ROWS")
print(data.head())

print("\nDATASET SHAPE")
print(data.shape)

# =========================================================
# q. DATA CLEANING
# =========================================================

# Replace ? with NaN

data.replace('?', np.nan, inplace=True)

# Convert all columns to numeric

for column in columns:

    data[column] = pd.to_numeric(

        data[column],

        errors='coerce'

    )

# Remove missing values

data.dropna(inplace=True)

# Remove duplicate rows

data.drop_duplicates(inplace=True)

# Remove negative values

for column in columns:

    data = data[data[column] >= 0]

print("\nDATA AFTER CLEANING")
print(data.shape)

# =========================================================
# r. OUTLIER DETECTION
# =========================================================

print("\nOUTLIER DETECTION")

for column in data.drop('CLASS', axis=1).columns:

    Q1 = data[column].quantile(0.25)

    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    outliers = data[

        (data[column] < lower)

        |

        (data[column] > upper)

    ]

    print(

        "Outliers in",

        column,

        ":",

        len(outliers)

    )

# =========================================================
# s. DATA TRANSFORMATION
# =========================================================

# Convert target values
# 1 = DIE -> 0
# 2 = LIVE -> 1

data['CLASS'] = data['CLASS'].replace({

    1: 0,

    2: 1

})

# Separate input and output

X = data.drop('CLASS', axis=1)

Y = data['CLASS']

print("\nCLASS VALUE COUNTS")
print(Y.value_counts())

# =========================================================
# FEATURE SCALING
# =========================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# =========================================================
# TRAIN TEST SPLIT
# =========================================================

X_train, X_test, Y_train, Y_test = train_test_split(

    X_scaled,

    Y,

    test_size=0.20,

    random_state=42

)

print("\nTRAINING DATA SIZE")
print(X_train.shape)

print("\nTESTING DATA SIZE")
print(X_test.shape)

# =========================================================
# t. LOGISTIC REGRESSION MODEL
# =========================================================

logistic_model = LogisticRegression(

    max_iter=1000

)

# Train model

logistic_model.fit(

    X_train,

    Y_train

)

# Prediction

logistic_prediction = logistic_model.predict(

    X_test

)

# Accuracy

logistic_accuracy = accuracy_score(

    Y_test,

    logistic_prediction

)

print("\n===================================")
print("LOGISTIC REGRESSION RESULT")
print("===================================")

print(

    "Accuracy =",

    round(logistic_accuracy * 100, 2),

    "%"

)

# =========================================================
# NAIVE BAYES MODEL
# =========================================================

naive_model = GaussianNB()

# Train model

naive_model.fit(

    X_train,

    Y_train

)

# Prediction

naive_prediction = naive_model.predict(

    X_test

)

# Accuracy

naive_accuracy = accuracy_score(

    Y_test,

    naive_prediction

)

print("\n===================================")
print("NAIVE BAYES RESULT")
print("===================================")

print(

    "Accuracy =",

    round(naive_accuracy * 100, 2),

    "%"

)

# =========================================================
# MODEL COMPARISON
# =========================================================

print("\n===================================")
print("MODEL COMPARISON")
print("===================================")

print(

    "Logistic Regression Accuracy :",

    round(logistic_accuracy * 100, 2),

    "%"

)

print(

    "Naive Bayes Accuracy :",

    round(naive_accuracy * 100, 2),

    "%"

)

if logistic_accuracy > naive_accuracy:

    print(

        "\nLogistic Regression performs better."

    )

elif naive_accuracy > logistic_accuracy:

    print(

        "\nNaive Bayes performs better."

    )

else:

    print(

        "\nBoth models perform equally."

    )

# =========================================================
# END OF PROGRAM
# =========================================================
```
