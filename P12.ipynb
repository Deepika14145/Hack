```python
import pandas as pd

# -------------------------------------------------
# Correct Column Names
# -------------------------------------------------

columns = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education_num',
    'marital_status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital_gain',
    'capital_loss',
    'hours_per_week',
    'native_country',
    'income'
]

# -------------------------------------------------
# Load Dataset Correctly
# header=None because dataset has no headers
# -------------------------------------------------

df = pd.read_csv(
    "adult.csv",
    header=None,
    names=columns,
    skipinitialspace=True
)

# -------------------------------------------------
# Display Data
# -------------------------------------------------

print("First 5 Rows:")
print(df.head())

print("\nColumn Names:")
print(df.columns)

# =================================================
# s. Create Data Subsets
# =================================================

# Subset for India
india_subset = df[df['native_country'] == 'India']

# Subset for Male
male_subset = df[df['sex'] == 'Male']

# Subset for White race
white_subset = df[df['race'] == 'White']

print("\nIndia Subset:")
print(india_subset.head())

print("\nMale Subset:")
print(male_subset.head())

print("\nWhite Race Subset:")
print(white_subset.head())

# =================================================
# t. Merge Two Subsets
# =================================================

merged_subset = pd.merge(
    india_subset,
    male_subset,
    how='inner'
)

print("\nMerged Subset:")
print(merged_subset.head())

# =================================================
# u. Sort Data
# =================================================

sorted_data = df.sort_values(
    by=['age', 'hours_per_week'],
    ascending=[False, False]
)

print("\nSorted Data:")
print(sorted_data.head())

# =================================================
# v. Transpose Data
# =================================================

transpose_data = df.head().T

print("\nTransposed Data:")
print(transpose_data)

# =================================================
# w. Melt Data to Long Format
# =================================================

melt_data = pd.melt(
    df.head(),
    id_vars=['sex'],
    value_vars=['age', 'hours_per_week'],
    var_name='Attribute',
    value_name='Value'
)

print("\nMelted Data:")
print(melt_data)

# =================================================
# x. Cast Data to Wide Format
# =================================================

wide_data = df.pivot_table(
    index='sex',
    columns='race',
    values='hours_per_week',
    aggfunc='mean'
)

print("\nWide Format Data:")
print(wide_data)
```
