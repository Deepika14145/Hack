```python
# =========================================================
# BREAST CANCER PREDICTION PROJECT
# =========================================================
# Operations Performed:
# 1. Data Cleaning
# 2. Outlier Detection and Correction
# 3. Data Transformation
# 4. Logistic Regression Model
# 5. Naive Bayes Model
# 6. Accuracy Comparison
# =========================================================


# =========================================================
# STEP 1 : IMPORT REQUIRED LIBRARIES
# =========================================================

import pandas as pd
import numpy as np

# Machine Learning Libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Models

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

# Accuracy Metric

from sklearn.metrics import accuracy_score


# =========================================================
# STEP 2 : LOAD BREAST CANCER DATASET
# =========================================================

# Read CSV file

data = pd.read_csv(
    "BreastCancerWc.csv",
    header=None
)

print("ORIGINAL DATASET SHAPE")
print(data.shape)


# =========================================================
# STEP 3 : ADD COLUMN NAMES
# =========================================================

data.columns = [

    'SampleCodeNumber',
    'ClumpThickness',
    'UniformityCellSize',
    'UniformityCellShape',
    'MarginalAdhesion',
    'SingleEpithelialCellSize',
    'BareNuclei',
    'BlandChromatin',
    'NormalNucleoli',
    'Mitoses',
    'Class'
]

# Display first 5 rows

print("\nFIRST 5 ROWS")
print(data.head())

# Dataset shape

print("\nDATASET SHAPE")
print(data.shape)

# Display column names

print("\nCOLUMN NAMES")
print(data.columns)


# =========================================================
# STEP 4 : DATA CLEANING
# =========================================================

# -----------------------------------------
# Replace '?' with NaN
# -----------------------------------------

data.replace('?', np.nan, inplace=True)

# -----------------------------------------
# Convert all columns to numeric
# -----------------------------------------

for column in data.columns:

    data[column] = pd.to_numeric(
        data[column],
        errors='coerce'
    )

# -----------------------------------------
# Remove rows having missing values
# -----------------------------------------

data.dropna(inplace=True)

# -----------------------------------------
# Remove duplicate rows
# -----------------------------------------

data.drop_duplicates(inplace=True)

# -----------------------------------------
# Remove negative values
# -----------------------------------------

numeric_columns = data.select_dtypes(
    include=np.number
).columns

for column in numeric_columns:

    data = data[
        data[column] >= 0
    ]

# -----------------------------------------
# Display cleaned dataset
# -----------------------------------------

print("\nDATASET AFTER CLEANING")
print(data.shape)

# =========================================================
# STEP 5 : OUTLIER DETECTION AND CORRECTION
# =========================================================

# Columns used for outlier detection

outlier_columns = [

    'ClumpThickness',
    'UniformityCellSize',
    'UniformityCellShape',
    'MarginalAdhesion',
    'SingleEpithelialCellSize',
    'BareNuclei',
    'BlandChromatin',
    'NormalNucleoli',
    'Mitoses'
]

# Apply IQR Method

for column in outlier_columns:

    Q1 = data[column].quantile(0.25)

    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR

    upper_limit = Q3 + 1.5 * IQR

    data = data[

        (data[column] >= lower_limit) &
        (data[column] <= upper_limit)
    ]

print("\nDATASET AFTER OUTLIER REMOVAL")
print(data.shape)

# Check remaining classes

print("\nCLASS COUNTS AFTER OUTLIER REMOVAL")
print(data['Class'].value_counts())

# =========================================================
# STEP 6 : DATA TRANSFORMATION
# =========================================================

# Target column

target_column = 'Class'

# Convert:
# 2 = Benign = 0
# 4 = Malignant = 1

data[target_column] = data[target_column].apply(

    lambda x: 0 if x == 2 else 1
)

# Separate input and output

X = data.drop(target_column, axis=1)

Y = data[target_column]

# Display target values

print("\nTARGET VALUE COUNTS")
print(Y.value_counts())

# -----------------------------------------
# Feature Scaling
# -----------------------------------------

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)


# =========================================================
# STEP 7 : SPLIT DATA INTO TRAINING AND TESTING
# =========================================================

X_train, X_test, Y_train, Y_test = train_test_split(

    X_scaled,
    Y,

    test_size=0.20,

    random_state=42
)

print("\nTRAINING DATA SIZE")
print(X_train.shape)

print("\nTESTING DATA SIZE")
print(X_test.shape)


# =========================================================
# STEP 8 : LOGISTIC REGRESSION MODEL
# =========================================================

# Create model

logistic_model = LogisticRegression(
    max_iter=1000
)

# Train model

logistic_model.fit(
    X_train,
    Y_train
)

# Predict output

logistic_prediction = logistic_model.predict(
    X_test
)

# Calculate accuracy

logistic_accuracy = accuracy_score(

    Y_test,
    logistic_prediction
)

print("\n===================================")
print("LOGISTIC REGRESSION RESULT")
print("===================================")

print(
    "Accuracy =",
    round(logistic_accuracy * 100, 2),
    "%"
)


# =========================================================
# STEP 9 : NAIVE BAYES MODEL
# =========================================================

# Create Naive Bayes model

naive_model = GaussianNB()

# Train model

naive_model.fit(
    X_train,
    Y_train
)

# Predict output

naive_prediction = naive_model.predict(
    X_test
)

# Calculate accuracy

naive_accuracy = accuracy_score(

    Y_test,
    naive_prediction
)

print("\n===================================")
print("NAIVE BAYES RESULT")
print("===================================")

print(
    "Accuracy =",
    round(naive_accuracy * 100, 2),
    "%"
)


# =========================================================
# STEP 10 : MODEL COMPARISON
# =========================================================

print("\n===================================")
print("MODEL COMPARISON")
print("===================================")

print(
    "Logistic Regression Accuracy :",
    round(logistic_accuracy * 100, 2),
    "%"
)

print(
    "Naive Bayes Accuracy :",
    round(naive_accuracy * 100, 2),
    "%"
)

# Compare models

if logistic_accuracy > naive_accuracy:

    print(
        "\nLogistic Regression performs better."
    )

elif naive_accuracy > logistic_accuracy:

    print(
        "\nNaive Bayes performs better."
    )

else:

    print(
        "\nBoth models perform equally."
    )


# =========================================================
# END OF PROJECT
# =========================================================
```
