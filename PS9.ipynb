```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep=';')
df
```

```python
df.head(5)
```

```python
# Perform the following operations using Python on the Facebook metrics data sets
# a. Create data subsets for type of post
# b. Merge two subsets
# c. Sort Data on Page total likes
# d. Transposing Data
# e. Melting Data to long format
# f. Casting data to wide format
```

```python
subset_photo = df[df['Type'] == 'Photo']
subset_video = df[df['Type'] == 'Video']
```

```python
print("\nPhoto Posts: ")
print(subset_photo.head())
```

```python
print("\nVideo Posts: ")
print(subset_video.head())
```

```python
#merge two subsets
merged_subset = pd.concat([subset_photo, subset_video])

print("\nMerge Data: ")
print(merged_subset.head())
```

```python
#sort data on page total likes
sorted_df = df.sort_values(by='Page total likes', ascending= False)

print("\nSorted Data: ")
print(sorted_df.head(50))
```

```python
#transpose data
transposed_df = df.transpose()

print(transposed_df.head())
```

```python
#melting 
melted_df = pd.melt(df,
                    id_vars=['Type'],
                    value_vars=['like', 'comment', 'share'],
                    var_name='Metric',
                    value_name='Value'
                   )
```

```python
print("\nMelted Data: ")
print(melted_df.head())
```

```python
#casting
wide_df = melted_df.pivot_table(
    index='Type',
    columns='Metric',
    values='Value',
    aggfunc='mean'
)
```
