```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv', header=None, names=[
    'SepalLength',
    'SepalWidth',
    'PetalLength',
    'PetalWidth',
    'Species'
    ]
)
df
```

```python
print("\nColumns in daaset: \n")
print(df.head())
```

```python
#create data subset for different species

setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']

print("\nSetosa Data:\n")
print(setosa.head())

print("\nVersicolor Data:\n")
print(versicolor.head())

print("\nVirginica Data:\n")
print(virginica.head())
```

```python
#merge two subsets
merged_data = pd.concat([setosa, versicolor])
print("\nMerged data: ")
print(merged_data.head(100))
```

```python
#sort data on the basis of petal length
sort_data = df.sort_values(by='PetalLength', ascending= True)
print("\nSorted data: ")
print(sort_data.head())
```

```python
#transpose data
transposed_data = df.transpose()
print("\nTransposed data: ")
print(transposed_data.head())
```

```python
#melting data to long format
melted_data = pd.melt(df,
                      id_vars=['Species'],
                      value_vars=['SepalLength',
                                  'SepalWidth',
                                  'PetalLength',
                                  'PetalWidth'
                                 ],
                      var_name='Measurement',
                      value_name='Value'
                     )

print("\nMleted Data: ")
print(melted_data.head())
```

```python
#casting data to wide format
wide_data = melted_data.pivot_table(
    index='Species',
    columns='Measurement',
    values='Value',
    aggfunc='mean'
)
print("\nWide format data:")
print(wide_data)
```
