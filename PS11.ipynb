```python
import pandas as pd

# =====================================================
# Load Datasets
# =====================================================

movies = pd.read_csv("movies_metadata.csv", low_memory=False)
ratings = pd.read_csv("ratings_small.csv")

# =====================================================
# Data Preprocessing
# =====================================================

# Convert movie id into numeric datatype
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')

# Remove null values
movies = movies.dropna(subset=['id'])

# Convert id to integer
movies['id'] = movies['id'].astype(int)

# =====================================================
# m) Create Data Subsets for Different Languages
# =====================================================

english = movies[movies['original_language'] == 'en']
french = movies[movies['original_language'] == 'fr']
hindi = movies[movies['original_language'] == 'hi']

print("English Movies:")
print(english[['title', 'original_language']].head())

print("\nFrench Movies:")
print(french[['title', 'original_language']].head())

print("\nHindi Movies:")
print(hindi[['title', 'original_language']].head())

# =====================================================
# n) Merge Two Subsets using concat()
# =====================================================

merged_data = pd.concat([english, french], ignore_index=True)

print("\nMerged Data:")
print(merged_data[['title', 'original_language']].head(10))

# =====================================================
# Merge with Ratings Dataset
# =====================================================

final_data = pd.merge(
    merged_data,
    ratings,
    left_on='id',
    right_on='movieId',
    how='inner'
)

print("\nData after merging with ratings:")
print(final_data[['title', 'rating']].head())

# =====================================================
# o) Sort Data using Customer Ratings
# =====================================================

sorted_data = final_data.sort_values(
    by='rating',
    ascending=False
)

print("\nTop Rated Movies:")
print(sorted_data[['title', 'original_language', 'rating']].head(10))

# =====================================================
# p) Transposing Data
# =====================================================

sample_data = sorted_data[
    ['title', 'original_language', 'rating']
].head(5)

transposed_data = sample_data.T

print("\nTransposed Data:")
print(transposed_data)

# =====================================================
# q) Melting Data to Long Format
# =====================================================

sample_df = pd.DataFrame({
    'Movie': ['Movie1', 'Movie2'],
    'Action_Rating': [4.5, 4.0],
    'Drama_Rating': [3.5, 4.8]
})

melted_data = pd.melt(
    sample_df,
    id_vars=['Movie'],
    var_name='Genre',
    value_name='Rating'
)

print("\nMelted Data (Long Format):")
print(melted_data)

# =====================================================
# r) Casting Data to Wide Format
# =====================================================

wide_data = melted_data.pivot(
    index='Movie',
    columns='Genre',
    values='Rating'
)

print("\nWide Format Data:")
print(wide_data)

# =====================================================
# Workflow Completed
# =====================================================

print("\nAll Operations Completed Successfully")
```
