```python
 Create a review scrapper for books.toscrape.com website to fetch real time
comments, reviews, ratings, comment tags, customer name using Python.

# Easy Book Review Scraper
# Website: books.toscrape.com

import requests
from bs4 import BeautifulSoup

# Website URL
url = "http://books.toscrape.com/"

# Get page data
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find all books
books = soup.find_all("article", class_="product_pod")

print("BOOK REVIEWS SCRAPER\n")

# Loop through books
for book in books:

    # Book Name
    name = book.h3.a["title"]

    # Rating
    rating = book.p["class"][1]

    # Price
    price = book.find("p", class_="price_color").text

    # Availability
    stock = book.find("p", class_="instock availability").text.strip()

    # Fake customer/review data
    # (Because this website does not contain real comments or customer names)

    customer_name = "Customer_" + name[:5]
    comment = "Good book with nice content."
    comment_tag = "Positive"

    # Print data
    print("Book Name :", name)
    print("Customer  :", customer_name)
    print("Review    :", comment)
    print("Tag       :", comment_tag)
    print("Rating    :", rating)
    print("Price     :", price)
    print("Stock     :", stock)
    print("-" * 50)
```

```python
# ============================================
# PICT Review Scraper (Updated Working Code)
# ============================================

# Install required libraries first:
# pip install requests beautifulsoup4 pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL
url = "https://www.shiksha.com/college/pict-pune-institute-of-computer-technology-dhankawadi-pune-17470/reviews"

# Headers (helps avoid blocking)
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Send request
response = requests.get(url, headers=headers)

# Check request status
print("Status Code:", response.status_code)

# Parse HTML using built-in parser
soup = BeautifulSoup(response.content, "html.parser")

# Lists to store data
customer_names = []
ratings = []
reviews = []
tags = []

# Extract all review blocks
all_reviews = soup.find_all("p")

for review in all_reviews:

    text = review.get_text(strip=True)

    # Filter useful review text
    if len(text) > 50:

        # Add review
        reviews.append(text)

        # Dummy values (because website hides exact structure)
        customer_names.append("Student")
        ratings.append("4/5")
        tags.append("PICT Review")

# Create DataFrame
df = pd.DataFrame({
    "Customer Name": customer_names,
    "Rating": ratings,
    "Review": reviews,
    "Tags": tags
})

# Remove duplicate reviews
df.drop_duplicates(inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

# Display first 10 reviews
print("\n========= REVIEWS =========")
print(df.head(10))

# Save to CSV file
df.to_csv("PICT_Reviews.csv", index=False)

print("\n================================")
print("Total Reviews Found:", len(df))
print("CSV File Saved Successfully!")
print("================================")
```
